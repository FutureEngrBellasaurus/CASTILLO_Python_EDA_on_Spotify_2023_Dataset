```python
#Import the Spotify 2023 Dataset
spotifydata2023 = pd.read_csv('spotify-2023.csv')
#Convert to Data Frame for better viewing
spotifydata = pd.DataFrame(spotifydata2023, columns = [
        'track_name','artist(s)_name','artist_count','released_year',
        'released_month','released_day','in_spotify_playlists','in_spotify_charts','streams',
        'in_apple_playlists','in_apple_charts','in_deezer_playlists',
        'in_deezer_charts','in_shazam_charts','bpm','key','mode',
        'danceability_%','valence_%','energy_%','acousticness_%',
        'instrumentalness_%','liveness_%','speechiness_%'
        ])
spotifydata
```

### B. Basic Descriptive Statistics
### *1. Mean, median, and standard deviation of Streams column*
### *2. Distribution of released_year and artist_count and its noticeable trends and outliers*

```python
#Since the E in the streams column is being considered as string let's convert it first to scientfic notation
#Function to convert scientific notation string to float
def convert_scientific_notation(streams):
    if isinstance(streams, str) and 'E' in streams:
        base, exponent = streams.split('E')
        return float(base) * 10**int(exponent)
    else:
        return float(streams)

#Apply the function to the 'value' column
spotifydata['streams'] = spotifydata['streams'].apply(convert_scientific_notation)
streams_converted = spotifydata['streams']
print(streams_converted)

#Replace the streams column with converted streams column
spotifydata = spotifydata.assign(streams=streams_converted)
```

```python
#Mean of Streams column
mean_streams = spotifydata['streams'].mean()
print("Mean of Streams Column:",mean_streams)
#Median of Streams column
median_streams = spotifydata['streams'].median()
print("Median of Streams Column:",median_streams)
#Std. Deviation of Streams column
standarddev_streams = spotifydata['streams'].std()
print("Std. Deviation of Streams Column:",standarddev_streams)
```

```python
#Group the data by release year and count the number of songs
tracks_per_year = spotifydata.groupby('released_year').size()

#Create and customize the bar plot
plt.figure(figsize=(15, 6))
tracks_per_year.plot(kind='bar', color='#f90093')
plt.xlabel('Released Year', color='#9000b3')
plt.ylabel('Number of Tracks', color='#43b929')
plt.title('Number of Tracks Released per Year', color='#735cdd')
plt.xticks(rotation=45, color='#9000b3')
plt.yticks(color='#43b929')
plt.grid(axis='y', alpha=1, color='#43b929')

#Group the data by artist count and count the number of trck
artist_count_column = spotifydata.groupby('artist_count').size()
#Create and customize the bar plot
plt.figure(figsize=(15, 6))
artist_count_column.plot(kind='bar', color='#f90093')
plt.xlabel('Artist Count', color='#9000b3')
plt.ylabel('Number of Tracks', color='#43b929')
plt.title('Distribution of Tracks by Artist Count', color='#735cdd')
plt.xticks(rotation=0, color='#9000b3')
plt.yticks(color='#43b929')
plt.grid(axis='y', alpha=1, color='#43b929')

plt.show()
```

### E. Genre and Music Characteristics
### *1. Correlation between streams and musial attributes (bpm, danceability_%, and energy_%) and observe which attribute influences streams the most.*
### *2. Correlation between danceability_% and energy_% and valence_% and acousticness_%.*

```python
#Correlation between STREAMS VS BPM
#Create and customize the scatter plot
plt.figure(figsize=(15, 6))
plt.scatter(spotifydata['bpm'], spotifydata['streams'], color='#f90093')
plt.xlabel('BPM', color='#9000b3')
plt.ylabel('Streams', color='#43b929')
plt.title('Correlation of Streams and BPM', color='#735cdd')
plt.xticks(rotation=0, color='#9000b3')
plt.yticks(color='#43b929')
plt.grid(axis='y', alpha=1, color='#43b929')
plt.show()
```

#### *STREAMS VS BPM:* The data points are all over the place, not forming a line. This shows there's no clear connection between how fast a song is (BPM) and how popular it is (streams). In other words, they have a weak correlation.

```python
#Correlation between STREAMS VS DANCEABILITY %
#Create and customize the scatter plot
plt.figure(figsize=(15, 6))
plt.scatter(spotifydata['danceability_%'], spotifydata['streams'], color='#f90093')
plt.xlabel('Danceability %', color='#9000b3')
plt.ylabel('Streams', color='#43b929')
plt.title('Correlation of Streams and Danceability %', color='#735cdd')
plt.xticks(rotation=0, color='#9000b3')
plt.yticks(color='#43b929')
plt.grid(axis='y', alpha=1, color='#43b929')
plt.show()
```

#### *STREAMS VS DANCEABILITY %*: As we can observe we can see that the danceability % scatter plots are lined up however this still doesn't signify a strong correlation between the said attribute and the streams because there is no linear relationship seen in the scatter plots.

```python
#Correlation between STREAMS VS ENERGY %
#Create and customize the scatter plot
plt.figure(figsize=(15, 6))
plt.scatter(spotifydata['energy_%'], spotifydata['streams'], color='#f90093')
plt.xlabel('Energy %', color='#9000b3')
plt.ylabel('Streams', color='#43b929')
plt.title('Correlation of Streams and Energy %', color='#735cdd')
plt.xticks(rotation=0, color='#9000b3')
plt.yticks(color='#43b929')
plt.grid(axis='y', alpha=1, color='#43b929')
plt.show()
```

#### *STREAMS VS ENERGY %:* Like the scatter plot of streams vs danceability %, as the energy of a song increases, the number of streams also tends to increase slightly. However, the relationship is not very strong, as there is a wide scatter of data points.

#### *CONCLUSION (STREAMS VS ATTRIBUTES):* Weak positive correlations may be seen in the scatter plots, suggesting that stream counts are influenced by more than just these characteristics. In contrast to BPM and energy, Danceability may have a marginally greater link with streams if we were to make a rough judgment based on the visual patterns. This is because a positive trend line seems to be more closely centered around the data points in the danceability plot, indicating a somewhat more stable relationship.

```python
#Correlation between DANCEABILITY % VS ENERGY %
#Create and customize the scatter plot
plt.figure(figsize=(15, 6))
plt.scatter(spotifydata['danceability_%'], spotifydata['energy_%'], color='#f90093')
plt.xlabel('Danceability %', color='#9000b3')
plt.ylabel('Energy %', color='#43b929')
plt.title('Correlation of Danceability % and Energy %', color='#735cdd')
plt.xticks(rotation=0, color='#9000b3')
plt.yticks(color='#43b929')
plt.grid(axis='y', alpha=1, color='#43b929')
plt.show()
```

#### *DANCEABILITY % VS ENERGY %:* It may seem like there is no perfectly linear correlation however, there appears to be a positive correlation between Danceability % and Energy %. This means that as the Danceability % increases, the Energy % also tends to increase. Songs that are more danceable are likely to have higher energy levels, according to the positive association. Given that danceable tunes frequently have a powerful rhythm and an energetic speed, this makes natural sense.

### G. Advanced Analysis
### *1. Based on the streams data, can you identify any patterns among tracks with the same key or mode (Major vs. Minor)?*
### *2. Do certain genres or artists consistently appear in more playlists or charts? Perform an analysis to compare the most frequently appearing artists in playlists or charts.*


```python
#Group the data by streams and key
tracks_key = spotifydata.groupby('key')['streams'].sum()

#Create and customize the bar plot
plt.figure(figsize=(15, 6))
tracks_key.plot(kind='bar', color='#f90093')
plt.xlabel('Key', color='#9000b3')
plt.ylabel('Number of Streams', color='#43b929')
plt.title('Number of Streams per Key', color='#735cdd')
plt.xticks(rotation=0, color='#9000b3')
plt.yticks(color='#43b929')
plt.grid(axis='y', alpha=1, color='#43b929')

#Group the data by streams and mode
tracks_mode = spotifydata.groupby('mode')['streams'].sum()

#Create and customize the bar plot
plt.figure(figsize=(15, 6))
tracks_mode.plot(kind='bar', color='#f90093')
plt.xlabel('Mode', color='#9000b3')
plt.ylabel('Number of Streams', color='#43b929')
plt.title('Number of Streams per Mode', color='#735cdd')
plt.xticks(rotation=0, color='#9000b3')
plt.yticks(color='#43b929')
plt.grid(axis='y', alpha=1, color='#43b929')
```

#### *Key and Major:* As observed in these bar charts, we can conclude that the mostly streamed songs have a C# key and a Major mode and this is because C major have no flats or sharps making the songs easier to be sung. Addition to that Major mode gives off positive emotions and upbeat moods.
